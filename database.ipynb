```python
import pandas as pd
from tqdm import tqdm

new_rows = []

for index, row in tqdm(df.iterrows(), total=len(df), desc="Processing rows"):
    start_year = row['startyear']
    end_year = row['endyear']
    
    years = range(start_year, end_year + 1)
    
    for year in years:
        new_row = {
            'year': year,
            'country': row['country'],
            'birth_year': row['birth_year'],
            'role_k150': row['role_k150'],
            'is_ai_broad': row['is_ai_broad'],
            'naics2': row['naics2']
        }
        
        new_rows.append(new_row)

result_df = pd.DataFrame(new_rows)
result_df.reset_index(drop=True, inplace=True)
```

```python
import pandas as pd
from tqdm import tqdm

grouped = result_df.groupby(['country', 'naics2', 'role_k150', 'year', 'age'])

results = []

for keys, group in tqdm(grouped, desc="Processing groups"):
    result = {
        'country': keys[0],
        'naics2': keys[1],
        'role_k150': keys[2],
        'year': keys[3],
        'age': keys[4],
        'total_count': len(group),
        'ai_broad_count': (group['is_ai_broad'] == 1).sum()
    }
    results.append(result)

final_df = pd.DataFrame(results)

final_df.to_csv('E:/nature aging/age_level_aggregated_result.csv', index=False)
final_df
```

```python
import pandas as pd
from tqdm import tqdm

bins = [15, 19, 24, 29, 34, 39, 44, 49, 54, 59, 64, 69, 74, 80]
labels = ['15-19', '20-24', '25-29', '30-34', '35-39', '40-44',
          '45-49', '50-54', '55-59', '60-64', '65-69', '70-74', '75+']

result_df['age_group'] = pd.cut(result_df['age'], bins=bins, labels=labels, right=False)

tqdm.pandas(desc="Grouping and aggregating")

final_df = result_df.groupby(['country', 'naics2', 'role_k150', 'year']).agg(
    total_count=('is_ai_broad', 'size'),
    ai_broad_count=('is_ai_broad', lambda x: (x == 1).sum()),
    **{f'{label}_count': ('age_group', lambda x, l=label: (x == l).sum()) for label in labels},
    **{f'{label}_ai_broad_count': (
        'age_group', 
        lambda x, l=label: result_df.loc[x.index, 'is_ai_broad'][(x == l)].sum()
    ) for label in labels}
).reset_index()

final_df.to_csv('E:/nature aging/aggregated_result.csv', index=False)

final_df
```

```python
import pandas as pd

age_thresholds = [35, 40, 45, 50, 55, 60, 65]

def compute_stats(group):
    total = len(group)
    ai_group = group[group['is_ai_broad'] == 1]
    ai_total = len(ai_group)

    result = {
        'total_count': total,
        'ai_total_count': ai_total,
        'ai_in_total': ai_total / total if total > 0 else 0,

        'age_mean': group['age'].mean(),
        'age_std': group['age'].std(),
        'age_se': group['age'].std() / total**0.5 if total > 0 else 0,
        'age_median': group['age'].median(),
        'age_max': group['age'].max(),
        'age_min': group['age'].min(),

        'ai_age_mean': ai_group['age'].mean(),
        'ai_age_std': ai_group['age'].std(),
        'ai_age_se': ai_group['age'].std() / ai_total**0.5 if ai_total > 0 else 0,
        'ai_age_median': ai_group['age'].median(),
        'ai_age_max': ai_group['age'].max(),
        'ai_age_min': ai_group['age'].min(),
    }

    for age in age_thresholds:
        age_mask = group['age'] >= age
        ai_mask = ai_group['age'] >= age

        result.update({
            f'age{age}_count': age_mask.sum(),
            f'age{age}_in_total': age_mask.mean(),
            f'age{age}_ai_count': ai_mask.sum(),
            f'age{age}_ai_in_ai': ai_mask.mean() if ai_total > 0 else 0,
            f'age{age}_ai_in_total': ai_mask.sum() / total if total > 0 else 0
        })

    return pd.Series(result)

age_stats_df = result_df.groupby(['naics2', 'role_k150']).apply(compute_stats).reset_index()

age_stats_df.to_excel('E:/nature aging/age_ai_by_naics_role.xlsx', index=False)
age_stats_df
```
